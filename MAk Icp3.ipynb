```python
# Aggregating the data for 'Pulse' and 'Calories' using min, max, count, and mean
aggregation = df[['Pulse', 'Calories']].agg(['min', 'max', 'count', 'mean'])
print(aggregation)
```

```python
# Filter rows with calories between 500 and 1000
calories_filter = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print(calories_filter)
```

```python
# Filter rows with calories > 500 and pulse < 100
filter_calories_pulse = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print(filter_calories_pulse)
```

```python
# Create a new DataFrame without the 'Maxpulse' column
df_modified = df.drop('Maxpulse', axis=1)
print(df_modified)
```

```python
# Drop the 'Maxpulse' column from the main DataFrame
df.drop('Maxpulse', axis=1, inplace=True)
print(df)
```

```python
# Replace NaN values in the 'Calories' column with the mean
df['Calories'].fillna(df['Calories'].mean(), inplace=True)

# Convert the 'Calories' column to integer datatype
df['Calories'] = df['Calories'].astype(int)

# Check the new datatype of the 'Calories' column
print(df.dtypes)

# Display the first few rows to verify that the 'Calories' column is now in integer format
print("\nUpdated DataFrame with 'Calories' as integers:")
print(df.head())

```

```python
import matplotlib.pyplot as plt

# Create a scatter plot for 'Duration' and 'Calories'
df.plot(kind='scatter', x='Duration', y='Calories', color='blue')

# Display the plot
plt.title('Duration vs Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.show()
```
